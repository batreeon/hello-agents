```python
from tavily import TavilyClient
import os
from dotenv import load_dotenv
import json

# 加载环境变量
load_dotenv()
TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")
os.environ['TAVILY_API_KEY'] = TAVILY_API_KEY

tavily_client = TavilyClient(api_key=TAVILY_API_KEY)

response = tavily_client.search("Who is Leo Messi?")

print(json.dumps(response, indent=2, ensure_ascii=False))
```

```python
import httpx
from dotenv import load_dotenv
import json

# 加载环境变量
load_dotenv()
TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")

headers = {
    "Authorization": TAVILY_API_KEY,
}

response = httpx.post(
    "https://api.tavily.com/search",
    json={"query": "Who is Leo Messi?"},
    headers=headers,
)

print(json.dumps(response.json(), indent=2, ensure_ascii=False))
```

```python
from tavily import TavilyClient
from dotenv import load_dotenv

# 加载环境变量
load_dotenv()
TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")

tavily_client = TavilyClient(api_key=TAVILY_API_KEY)

response = tavily_client.search(
    query="latest trends in generative AI 2025",
    search_depth="advanced",
    topic="news",
    time_range="w",        # past week
    max_results=10,
    include_raw_content=True,
    include_images=False,
    include_answer=False,
    exclude_domains=["some-spammy-site.com","low-quality-blog.org"]
)

for res in response["results"]:
    print(res["title"], res["url"])
    print(res["content"][:200], "...")
```
